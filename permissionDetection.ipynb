```python

from androguard.core.bytecodes.apk import APK

def extract_permissions(apk_file):
    apk = APK(apk_file)
    permissions = apk.get_permissions()
    return permissions

# Create a set of all permissions
all_permissions = set()

# Find out all permissions
for apk_file in benign_apps + malicious_apps:
    apk = APK(apk_file)
    permissions = apk.get_permissions()
    all_permissions.update(permissions)
```

```python
import hashlib
import pandas as pd


# Function that calculates the hash value
def getHash(file_path):
    sha256 = hashlib.sha256()
    with open(file_path, 'rb') as f:
        for block in iter(lambda: f.read(65536), b''):
            sha256.update(block)
    return sha256.hexdigest()

# Create a dictionary where the data will be stored
data = {}

# Add permissions for benign apps
for apk_file in benign_apps:
    apk_hash = getHash(apk_file)
    apk_permissions = extract_permissions(apk_file)
    row = {permission: 1 if permission in apk_permissions else 0 for permission in all_permissions}
    row['class'] = 0
    data[apk_hash] = row

# Add permissions for malicious apps
for apk_file in malicious_apps:
    apk_hash = getHash(apk_file)
    apk_permissions = extract_permissions(apk_file)
    row = {permission: 1 if permission in apk_permissions else 0 for permission in all_permissions}
    row['class'] = 1
    data[apk_hash] = row

# Convert data to DataFrame
dataframe = pd.DataFrame(data).transpose()
dataframe = dataframe.fillna(0)

# Process DataFrame and show results
print(dataframe)
```

```python
# Print DataFrame to CSV file
dataframe.to_csv('dataset.csv', sep='\t')
```
